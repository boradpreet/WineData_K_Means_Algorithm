```python
import numpy as np  # use for Numerical Value
import pandas as pd  # use for Data Cleaning 
import matplotlib.pyplot as plt  # use for Data Visualization 
import seaborn as sns  # use for Data Visualization
```

```python
# Load the Dataset
data = pd.read_csv("wine-clustering.csv")
data
```

#### Domian Analysis:-

            * Alcohol : - Alcohol is what separates wine from grape juice. 
            * Malic_Acid :- Malic acid is an alpha hydroxy acid found in certain fruits and wines.
            * Ash :- Ash determination is an obligatory analysis for certified wines to be placed on the market.
            * Ash_Alcanity :- The alkalinity of the ash is defined as the sum of cations, other than the ammonium ion, combined 
                              with the organic acids  in the wine.
            * Magnesium :- Magnesium performs its most significant function as an essential component of chlorophyl.
            * Total_Phenols :- contributes to the astringency, bitterness, and other tactile sensations defined as structure or body,
                               as well as to the wine's red color.
            * Flavanoids :- Flavonoids include the anthocyanins and tannins which contribute to the color and mouthfeel of the wine.
            * Nonflavanoid_Phenols :- Phenolic compounds are secondary metabolites present in grapes and wine that can be formed and
                                      transformed during the winemaking process.
            * Proanthocyanins :- These condensed tannins strongly influence the perceived astringency of the wine.
            * Color_Intensity :- The more concentrated and opaque a wine's color, the higher its intensity.
            * Hue :- A purple or magenta tint indicates it's a very young wine, such as Beaujolais Nouveau (Gamay).
            * OD280 :- This method for calculating protein concentration can determine the amount of protein in various wines.
            * Proline :- Enhance viscosity, sweetness and flavour in red wines, while suppressing bitterness and astringency.

## Model Creation

```python
x = data[["Alcohol","Malic_Acid","Ash","Ash_Alcanity","Magnesium","Total_Phenols","Flavanoids","Nonflavanoid_Phenols","Proanthocyanins","Color_Intensity","Hue","OD280","Proline"]]
x
```

```python
from sklearn.cluster import KMeans # Import the Kmeans 
KM = KMeans(n_clusters=6) # assighn the cluster k value
KM.fit(x)  # Train the model
```

```python
# show the datapoint is which cluster
KM.labels_ 
```

```python
# create the six cluster and inside difference between datapoints to centroid 
KM.cluster_centers_
```

```python
# Create the Eelbow Technique
wcss = []
for i in range(1,11):
    KM = KMeans(n_clusters=i,random_state=40)
    KM.fit(x)
    KM.inertia_
    wcss.append(KM.inertia_)
```

```python
# create the Plot
plt.plot(range(1,11),wcss,marker = "^")
plt.show()
```

* Eelbow thechnique throw k value is 3

```python
from sklearn.cluster import KMeans # Import the Kmeans 
KM = KMeans(n_clusters=3) # assighn the cluster k value
KM.fit(x)  # Train the model
```

```python
# show the datapoint is which cluster
KM.labels_ 
```

```python
# create the six cluster and inside difference between datapoints to centroid 
KM.cluster_centers_
```

```python
lable = KM.labels_ # store the label value in label variable
lable

from sklearn.metrics import silhouette_score # import the Kmeans metrics
score = silhouette_score(x,lable)
score
```

```python
color_sum = np.array(["red","blue","green","orange"])
plt.scatter(x.Hue,x.Proline,color = color_sum[lable])
```

#### Conclusion:-

        * Here we can see the clear four cluster in scatter plot.
        * Cluster Centroid k value is four.
        * silhouette_score is a 0.5472492065729442 it is basically show the model performence quality in dataset.
